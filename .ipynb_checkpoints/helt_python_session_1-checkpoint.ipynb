### Notebooks
Jupyter Notebook er ett "program" som lar deg skrive Python kode. Du kan tenke på det som SAS EG for python.

Den Store forskjellen mellom EG og Jupyter notebooks er at det bruker såkalte "Notebooks" som "programm" for og skrive koden. Disse notebokkene har en rekke funksjoner som er meget praktiske, men også ganske uvandte. 
- Det første som du ser her er at du kan skrive tekst og Kode i samme fil, men i seperate "bokser" (Såkalte celler, eller paragrapher). 
- Disse cellene can brukes til å dele opp koden din i biter som kan kjøres uavhengig av hverandre.
- Det neste er at når du skriver kode, kan du se resultatet av koden direkte under koden du skrev.
- Det er ingen log slik du er vandt til fra SAS. 
    - Man kan lage en slik log, med det er noe mer avansert enn vi skal snakke om her
    
Det systemet som moderniserings prosjektet jobber med som program for å lage og kjøre statistikk er også en Notebook bassert løsning.
    
Det er mye mer som kan sies om Notebooks men det kommer vi til å dekke mere omfattende i egne kurs.

```python
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data/Titanic.csv')
df.head()
```

```python
df.dtypes
```

```python
df = pd.read_csv(PATH+'Data\\train.csv', sep = ',', encoding ='utf-8', dtype={'Pclass':'str'})
df.dtypes
```

```python
df.describe()
```

```python
df[['Age', 'Fare']].describe()
```

sett inn noe om is null count

```python
by_class = df.groupby('Pclass')[['Age', 'Fare']].mean()
by_class.head()
```

```python
by_class = by_class.rename({'Age': 'Mean_age', 'Fare': 'Mean_fare'}, axis=1)
by_class.head()
```

```python
df2 = pd.merge(df, by_class, on='Pclass', how='left')
df2.head()
```

```python
df2['dist_age'] = df2['Age']-df2['Mean_age']
df2['dist_fare'] = df2['Fare']-df2['Mean_fare']
df2.head()
```

```python
Cherbourg = df2.loc[df2['Embarked']=='C']
Cherbourg.sample(n=8)
```

```python
plt.figure(figsize=(20,10))
plt.hist(df2['Age'],  color = 'g', edgecolor = 'black',
         bins = int(16))
plt.title('Distribusjon av alder på Titanic', fontsize = 24)
plt.xlabel('Alder', fontsize = 14)
plt.ylabel('Antall', fontsize = 14)
```

```python
x=df2['Age']
y=df2['Fare']
colors = {'1':'red', '2':'blue', '3':'green'}

plt.figure(figsize=(10,10))
plt.scatter(x, y, color=df2['Pclass'].apply(lambda x: colors[x]))
plt.title('Pris vs alder på Titanic', fontsize = 24)
plt.xlabel('Age', fontsize = 14)
plt.ylabel('Fare', fontsize = 14)
```

```python
x=df2['Age']
y=df2['Fare']
colors = {'1':'red', '2':'blue', '3':'green'}

plt.figure(figsize=(10,10))
plt.scatter(x, y, color=df2['Pclass'].apply(lambda x: colors[x]))
plt.ylim(0,100)
plt.title('Pris vs alder på Titanic', fontsize = 24)
plt.xlabel('Age', fontsize = 14)
plt.ylabel('Fare', fontsize = 14)
```
